```python
import random

suits = ['Hearts', 'Diamonds', 'Clubs', 'Spades']
ranks = ['2', '3', '4', '5', '6', '7', '8', '9', '10', 'Jack', 'Queen', 'King', 'Ace']
deck = [{'rank': rank, 'suit': suit} for suit in suits for rank in ranks]

random.shuffle(deck)

players = [[], [], [], []]
for i in range(len(deck)):
    players[i % 4].append(deck[i])

def display_hand(player):
    for card in player:
        print(card['rank'], 'of', card['suit'])

def play_round(starter):
    round_trick = []
    player_order = list(range(starter, 4)) + list(range(starter))
    lead_suit = players[starter][0]['suit']

    for i in player_order:
        player = players[i]
        matching_suit_cards = [card for card in player if card['suit'] == lead_suit]

        if matching_suit_cards:
            card = matching_suit_cards.pop(0)
            player.remove(card)
        else:
            card = player[0]
            player.remove(card)

        round_trick.append(card)
        print(f"Player {i + 1} plays {card['rank']} of {card['suit']}")

    winning_card = max(round_trick, key=lambda x: (ranks.index(x['rank']), suits.index(x['suit'])))
    winner = (starter + round_trick.index(winning_card)) % 4
    print(f"Player {winner + 1} wins the trick!\n")
    return winner

starter = players.index([player for player in players if {'rank': '2', 'suit': 'Clubs'} in player][0])
for round in range(13):
    print(f"------ Round {round + 1} ------")
    winner = play_round(starter)
    starter = winner

heart_count = [player.count({'rank': 'Ace', 'suit': 'Hearts'}) for player in players]
winner = heart_count.index(min(heart_count)) + 1
print(f"Player {winner} wins the game!")

for i, player in enumerate(players):
    print(f"Player {i + 1}'s hand:")
    display_hand(player)
    print()
```
